```python
# Replacing null values in Age column with mean value.
data.Age.fillna(data.Age.mean(), inplace=True)
data['Age'] = data['Age'].astype(int)
```

```python
data.head()
```

```python
# # replacing -1 in the rating columns with mean
data["Rating"].replace(-1.0,0.0, inplace= True)
data['Rating'] = data['Rating'].round(1)
# now to replace one missing value using mean value
data.Rating.fillna(data.Rating.mean(), inplace=True)
data['Rating'] = data['Rating'].round(1)
```

```python
# Cleaning Location column
data['Location'].replace(' ',',',regex=True,inplace=True)
data[["Location","Code"]]= data['Location'].str.rsplit(',', n=1, expand=True)
```

```python
data['Location'] = data['Location'].str.strip()
data['Code'] = data['Code'].str.strip()
```

```python
data['Location'].head()
```

```python
del data['Code']
data['Location'].replace(',',' ',regex=True,inplace=True)
```

```python
data['Location'].head()
```

```python
# Starting Index from 1
data["Index"]= data["Index"]+1
```

```python
# Changing -1 into False
data['Easy Apply'].replace('-1','FALSE',inplace=True)
```

```python
# changing -1 to 0 because year are not relatively similar to take a mean value and dont want to write null or NaN , which might change column datatype.
data['Established'].replace(-1,0,inplace=True)
```

```python
# Salary column
data['Salary'] = data['Salary'].replace({'\$':'','k':'000'},regex=True)
```
